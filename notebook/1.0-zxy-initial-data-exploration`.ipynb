```python
import os
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python

cwd = os.getcwd()  # Get the current working directory

def load_data():
    data_filename = "dataset_cybersecurity_michelle.csv"  
    data_path = os.path.join(cwd, '..', 'data','raw', data_filename)  # Construct the full path
    return pd.read_csv(data_path)

# load the dataset
df= load_data()
df.head()


```

```python
df.info()
df.isnull().sum()
df.describe()
```

```python
missing_value_df = df[df.isin([-1]).any(axis=1)]
missing_value_df.describe()
```

```python
# Create a new DataFrame 'cleaned_df' with -1 replaced by mode
cleaned_df = df.copy()  # Make a copy of the original DataFrame

# Replace -1 with the mean of each column
for col in cleaned_df.columns:
    if -1 in cleaned_df[col].values:
        mean_val = cleaned_df[col][cleaned_df[col] != -1].mean()  # Calculate the mean excluding -1
        cleaned_df[col] = cleaned_df[col].replace(-1, mean_val)
cleaned_df.describe()
```
